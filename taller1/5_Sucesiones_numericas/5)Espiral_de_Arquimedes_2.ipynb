```python
#creado por Nicolas Daza & Sebastian Hurtado
"""Espiral de Arquimedes:
        Transformacion polar a cartesiana"""
import numpy as np
import matplotlib.pyplot as plt
matriz=[]
circum=(20000*np.pi)
def archimedes(a:int , b:int,N:int):
    matriz=np.zeros((N,2))
    for i in range(0,len(matriz)):
        theta=circum/(i+1)
        r=a+(b*theta)
        x=r*(np.cos(theta))
        y=r*(np.sin(theta))
        matriz[i,0]=x
        matriz[i,1]=y
    return matriz
new=archimedes(0,1,20000)
plt.scatter(new[:,0],new[:,1])
plt.xlim(-25,25)
plt.ylim(-25,25)
plt.show()

```
